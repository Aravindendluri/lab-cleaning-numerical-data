```python

def variable(x):
    lst_cont=[]
    lst_disc=[]
    for i in x:
        if len(x[i].unique())>350:
            lst_cont.append(i)
        else:
            lst_disc.append(i)
    return x[lst_cont], x[lst_disc] 
```

```python
# seperating continuous, discrete 
continuous_df, discrete_df = variable(customer_df)
```
